# Visualization

```python
diwali.columns
```

```python
diwali["Gender"].value_counts()
```

```python
gender=sns.countplot(x="Gender",data=diwali,palette="Set3")
plt.plot()

for bar in gender.containers:
    gender.bar_label(bar)
```

```python
#Gender vs Amount
```

```python
GvsA=diwali.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values("Amount",ascending=False)
```

```python
plt.pie(GvsA["Amount"],labels=GvsA["Gender"],autopct="%.0f%%",shadow=True,colors=["Green","Red"])
plt.legend(title="Gender")
plt.show()
```

```python
sns.barplot(x="Gender",y="Amount",data=GvsA)
plt.plot()
```

```python
# Age
```

```python
diwali["Age Group"].value_counts()
```

```python
a1=sns.countplot(x="Age Group",data=diwali,hue="Gender",palette="bright")
for bars in a1.containers:
    a1.bar_label(bars)
```

```python
AvsA=diwali.groupby(["Age Group","Gender"],as_index=False)["Amount"].sum().sort_values("Amount",ascending=False)
```

```python
AvsA
```

```python
sns.barplot(x="Age Group",y="Amount",data=AvsA,hue="Gender")
plt.plot()
```

We can see that most number of shopping is done by Age Group of 26-35 and also the purchasing rate of the same group is the highest
